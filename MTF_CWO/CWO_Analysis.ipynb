```python
from matplotlib.image import NonUniformImage
from pylab import figure, cm
from matplotlib.colors import LogNorm

plt.figure()

xedges = np.linspace(-0.98,0.98,51)
yedges = np.linspace(-0.98,0.98,51)

# H, xedges, yedges = np.histogram2d(xx, yy, bins=500)
plt.hist2d(xx,yy, bins = [xedges,yedges], cmap=cm.jet, norm=LogNorm())
axis('image')
plt.xlabel('X [cm]')
plt.ylabel('Y [cm]')
cbar = plt.colorbar()
cbar.set_label('Counts')
```

```python
fig = plt.figure(figsize=(7, 3))
ax = fig.add_subplot(111, title='pcolormesh: actual edges',aspect='equal')

xedges = np.linspace(0,401.408,1025)

yedges = np.linspace(0,301.056,769)

H, xedges, yedges = np.histogram2d(xx, yy, bins=(xedges, yedges))

H = H.T 

im = NonUniformImage(ax, interpolation='bilinear')

xcenters = (xedges[:-1] + xedges[1:]) / 2

ycenters = (yedges[:-1] + yedges[1:]) / 2

im.set_data(xcenters, ycenters, H)

ax.images.append(im)

plt.show()

X, Y = np.meshgrid(xedges, yedges)

plt.pcolormesh(X, Y, H)
```
