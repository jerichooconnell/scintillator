```python
n = 0
outF = open("Crystal_CsI.txt", "w")
for xx in np.round(np.linspace(-250*0.01021,250*0.01021,501),decimals=5):
    for yy in np.round(np.linspace(-250*0.0102,250*0.0102,501),decimals=5):

# for xx in np.round(np.linspace(-5*0.01021,5*0.01021,11),decimals=5):
#     for yy in np.round(np.linspace(-5*0.0102,5*0.0102,11),decimals=5):
        
        # write line to output file
        outF.write(f's:Ge/CsI_{n}/Parent   = "SeptaGroup"\n'),
        outF.write(f's:Ge/CsI_{n}/Type = "TsCylinder"\n'),
        outF.write(f'd:Ge/CsI_{n}/RMax = 0.0051 mm\n'),
        outF.write(f'd:Ge/CsI_{n}/HL = 0.180 mm\n'),
        outF.write(f's:Ge/CsI_{n}/Material = "CsI"\n')
        outF.write(f'd:Ge/CsI_{n}/TransX = {xx} mm\n'),
        outF.write(f'd:Ge/CsI_{n}/TransY = {yy} mm\n\n')
        
        n += 1
        
        
outF.close()
```
